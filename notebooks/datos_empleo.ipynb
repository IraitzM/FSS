Podemos intentar buscar en otras fuentes información adicional que nos ayude a proyectar qué es lo que ha podido pasar en Donosti basado en datos de Gipuzkoa (por ejemplo).

Datos de Open Data Euskadi: https://opendata.euskadi.eus/catalogo/-/estadistica/paro-registrado-desde-1997/

```python
pd.read_csv("https://opendata.euskadi.eus/contenidos/estadistica/paro_historico/es_def/adjuntos/Dap_Historico.csv", encoding="latin-1", sep=";")
```

```python
datos_euskadi = pd.read_csv("https://opendata.euskadi.eus/contenidos/estadistica/paro_historico/es_def/adjuntos/Dap_Historico.csv", encoding="latin-1", sep=";", skiprows=1, header=0)
datos_euskadi
```

```python
datos_euskadi.describe()
```

```python
datos_euskadi[datos_euskadi["T.H."] == "GIPUZKOA"]
```

```python
datos_euskadi[datos_euskadi["Cod. Comarca a 11"] == 8.0]
```

```python
datos_euskadi[datos_euskadi["Nombre Municipio"] == "DONOSTIA-SAN SEBASTIAN"].head(5)
```

```python
datos_euskadi.info()
```

```python
datos_euskadi[datos_euskadi["Nombre Municipio"] == "DONOSTIA-SAN SEBASTIAN"][["Año","Trimestre", "Pobl. Activa","TOTAL PARADOS"]]
```

```python
datos_promedio = datos_euskadi[datos_euskadi["Nombre Municipio"] == "DONOSTIA-SAN SEBASTIAN"][["Año", "Pobl. Activa","TOTAL PARADOS"]].groupby("Año").agg("mean")
datos_promedio.reset_index(inplace=True)
datos_promedio
```

```python
datos_promedio.dtypes
```

```python
datos_promedio["Urtea"] = (datos_promedio["Año"]*1000).astype(int)
```

```python
datos_promedio["Langabezia Tasa"] = (datos_promedio["TOTAL PARADOS"]/datos_promedio["Pobl. Activa"])*100
datos_promedio
```

```python
datos_completos = pd.concat([ciudad[["Urtea","Langabezia Tasa"]], datos_promedio[["Urtea","Langabezia Tasa"]]])
datos_completos
```

```python
datos_completos = datos_completos.groupby("Urtea").agg("mean")
datos_completos.reset_index(inplace=True)
```

```python
plot_unemployment_rates(datos_completos);
```

### Web scrapping

```python
import requests

URL = "https://datosmacro.expansion.com/paro/espana/municipios/pais-vasco/guipuzcoa/donostia-san-sebastian"
page = requests.get(URL)
```

```python
import pprint

pprint.pprint(page.content[:250])
```

Toca inspeccionar la web para ver cómo obtener la tabla que nos interesa.

```python
!pip install beautifulsoup4 lxml
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, "html.parser")
```

```python
from io import StringIO

results = soup.find(id="tb1")
data_scrap = pd.read_html(StringIO(str(results)))[0]
```

```python
data_scrap
```

```python
data_scrap.info()
```

```python
data_scrap.iloc[0,0] = '2024'
data_scrap["Urtea"] = data_scrap.Fecha.astype(int)
data_scrap["Langabezia Tasa"] = 2*(data_scrap["Nº de parados registrados"]/data_scrap["Población"])*100
data_scrap
```

```python
data_futura = data_scrap[data_scrap["Urtea"].isin([2024,2023,2022])]
data_futura
```

```python
datos_completos = pd.concat([ciudad[["Urtea","Langabezia Tasa"]], datos_promedio[["Urtea","Langabezia Tasa"]], data_futura[["Urtea","Langabezia Tasa"]]])
datos_completos
```

```python
datos_completos = datos_completos.groupby("Urtea").agg("mean")
datos_completos.reset_index(inplace=True)
```

```python
plot_unemployment_rates(datos_completos);
```
