Leemos el fichero de datos.

```python
import pandas as pd

fichero = "https://raw.githubusercontent.com/IraitzM/FSS/refs/heads/main/data/store.csv"

data = pd.read_csv(fichero, sep=";")
data
```

Queremos entender en qué campos somos mejores o peores al resto de tiendas.

```python
data["perf"] = data["ours"] - data["others"]
data
```

```python
data["perf"].plot();
```

```python
from visuals import plot_pos_neg_bars

plot_pos_neg_bars(data['perf'], data['offer'], "¿En qué somos buenos?");
```
